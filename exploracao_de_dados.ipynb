### Explorção das classes presentes na base de dados

```python
df_credit.Rating.value_counts()
```

```python
# Criar um histograma para visualizar a distribuição das classificações de crédito
rating_order = [
    "AAA", "AA", "A",
    "BBB", "BB", "B",
    "CCC", "CC", "C", "D"
]

fig = px.histogram(
    df_credit,
    x="Rating",
    category_orders={"Rating": rating_order}
)
fig.show()

# O grafico segue uma distribuição normal.
```

### Exlorando a distribuição de algumas features 

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=[
        "Current Ratio",
        "Quick Ratio",
        "Net Profit Margin",
        "Gross Profit Margin"
    ]
)

fig.add_trace(
    go.Box(y=df_credit["currentRatio"], name="Current Ratio"),
    row=1, col=1
)

fig.add_trace(
    go.Box(y=df_credit["quickRatio"], name="Quick Ratio"),
    row=1, col=2
)

fig.add_trace(
    go.Box(y=df_credit["netProfitMargin"], name="Net Profit Margin"),
    row=2, col=1
)

fig.add_trace(
    go.Box(y=df_credit["grossProfitMargin"], name="Gross Profit Margin"),
    row=2, col=2
)

fig.show()
```

A distribuição dos dados aponta para a presença signifcante de outliers. No mundo real, acredito que o dataset se beneficiaria de uma analise especializada
para definir se os outliers são erros de entrada ou dgtação ou se realmente represtam valores reais.

Uma forma de avaliar visualmente seria plotar cada feature a ser analisada separada por Rating:

```python
import plotly.express as px

rating_order = [
    "AAA", "AA", "A",
    "BBB", "BB", "B",
    "CCC", "CC", "C", "D"
]

fig = px.box(
    df_credit,
    x="Rating",
    y="netProfitMargin",
    category_orders={"Rating": rating_order}
)

fig.update_layout(
    title="Distribution of Net Profit Margin by Credit Rating",
    xaxis_title="Credit Rating",
    yaxis_title="Net Profit Margin"
)

fig.show()
```

### Outra possibilidade é analise temporal da evolução de scores, ratings e features de interesse:

```python
df_credit.groupby("Name")['Date'].count().sort_values(ascending=False)
```

```python
# Plot de pontos usando plotly como uma empresa específica tem o seu rating e net profit ao longo do tempo
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=[
        "Rating",
        "Net Profit Margin",
    ]
)

# Add traces
fig.add_trace(go.Scatter(x=df_credit[df_credit["Name"] == "Bunge Limited"]["Date"], y=df_credit[df_credit["Name"] == "Bunge Limited"]["Rating"],
                    mode='markers',
                    name='Rating'), row=1, col=1)
fig.add_trace(go.Scatter(x=df_credit[df_credit["Name"] == "Bunge Limited"]["Date"], y=df_credit[df_credit["Name"] == "Bunge Limited"]["grossProfitMargin"],
                    mode='markers',
                    name='grossProfitMargin'), row=2, col=1)

fig.show()
```

### Plot de correlação das variáveis numericas do dataset

```python

rating_dict = {'AAA': 0,
             'AA': 1,
             'A': 2,
             'BBB': 3,
             'BB': 4,
             'B': 5,
             'CCC': 6,
             'CC': 7,
             'C': 8,
             'D': 9}
# Map the 'Rating' column to the new 'Rating_id' column using the rating_dict
df_credit["Rating_id"] = df_credit["Rating"].map(rating_dict)
num_df = df_credit.select_dtypes(include=["int64", "float64"])
corr_spearman = num_df.corr(method="spearman")
```

```python
fig = px.imshow(
    corr_spearman,
    color_continuous_scale="RdBu",
    zmin=-1,
    zmax=1,
    title="Correlation Matrix (Numerical Features)"
)

fig.update_layout(
    width=900,
    height=900
)

fig.show()
```

Com a matriz de correlação podemos ver algumas features com alta correlação direta ou inversa mas nenhuma com alta correlação (>0.5) com nossa variável de interesse (Rating_id)
