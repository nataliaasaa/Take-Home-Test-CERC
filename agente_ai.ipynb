```python
df = pd.read_csv("data/credit_risk_prediction.csv")

display(df.head(3))

# Exibe empresas dispon√≠veis
print("\nüè¢ Empresas dispon√≠veis:")
for nome in sorted(df["Name"].unique()):
    print(f"  ‚Ä¢ {nome}")
```

```python
# Selecione a empresa (ou deixe None para an√°lise do portf√≥lio completo)
empresa = "Albany International Corporation" 

context = build_risk_context(df, empresa if empresa != "Todas" else None)

print("üìå Contexto gerado:")
print("="*70)
print(context)
print("="*70)
```

```python
# Inicializa modelo com o contexto embutido na instru√ß√£o do sistema
model = genai.GenerativeModel(
    model_name="gemini-flash-latest",
    system_instruction=f"""
Voc√™ √© um analista s√™nior de risco de cr√©dito corporativo.

REGRAS ESTRITAS:
1. USE APENAS os dados do contexto abaixo. NUNCA invente n√∫meros ou m√©tricas.
2. Se informa√ß√£o n√£o existir: "‚ö†Ô∏è Dado n√£o dispon√≠vel na an√°lise realizada."
3. Responda para comit√™ de cr√©dito: destaque riscos, drivers e recomenda√ß√µes.
4. Formato:
   - üìå Resumo executivo
   - üîç An√°lise detalhada
   - üí° Recomenda√ß√£o pr√°tica

CONTEXTO:
{context}
    """,
    generation_config=genai.types.GenerationConfig(
        temperature=0.3,
        max_output_tokens=2048
    )
)

chat = model.start_chat(history=[])
print(f"‚úÖ Modelo '{model.model_name}' pronto para an√°lise")
```

```python
# ‚¨áÔ∏è ALTERE ESTA PERGUNTA E EXECUTE A C√âLULA ‚¨áÔ∏è
pergunta = "Prepare a report that include the degree of risk involved in extending credit or lending money to Albany International Corporation."

# Envia pergunta e exibe resposta
print(f"\nüë§ Voc√™: {pergunta}\n")
print("ü§ñ Analista de Risco:\n")

response = chat.send_message(pergunta)
display(Markdown(response.text))

# Dica: Execute esta c√©lula novamente com outra pergunta para continuar o chat
```
