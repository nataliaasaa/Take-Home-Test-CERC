### Divide .csv file into train (for train and test) and evaluation

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# =========================
# Configurações
# =========================

INPUT_PATH = "./corporate_rating.csv"
TRAIN_PATH = "./corporate_rating_train.csv"
EVAL_PATH = "./corporate_rating_evaluation.csv"

TARGET_COL = "Rating"   # ou "Rating_id_encoded", se preferir

TEST_SIZE = 0.2
RANDOM_STATE = 42

# =========================
# Load data
# =========================

df = pd.read_csv(INPUT_PATH)

print(f"Dataset original: {df.shape}")

# =========================
# Split estratificado
# =========================

df_train, df_eval = train_test_split(
    df,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE
)

print(f"Train shape: {df_train.shape}")
print(f"Evaluation shape: {df_eval.shape}")

# =========================
# Save files
# =========================

df_train.to_csv(TRAIN_PATH, index=False)
df_eval.to_csv(EVAL_PATH, index=False)

print("Arquivos salvos com sucesso:")
print(f" - {TRAIN_PATH}")
print(f" - {EVAL_PATH}")
```
